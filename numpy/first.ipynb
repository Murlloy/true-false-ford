```python
#EXERCICE 4

import numpy as np

array = np.array([100,200,300,400,500,600])

print(f"Matriz Original: {array}\n")

array_1, array_2 = np.split(array, 2)

print(f"Matriz Redimencionada: \n\n{array_1}\n{array_2}\n")

array_2 = np.append(array_2, [700])

print(f"Matriz Modificada: \n\n{array_1}\n{array_2}")
```
